```python
import pandas as pd
```

```python
import nltk as nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid=  SentimentIntensityAnalyzer()
```

```python
import numpy as np
```

```python
df=pd.read_csv('Amazon_reviews.csv',index_col=['Id'])
df.head(900)
```

```python
# first of all delete column 
df.drop(columns=['Price','ProductId','UserId','Helpfulness','Time','Text'], inplace=True)
df
```

```python
df.drop(columns=['ProfileName'], inplace=True)
df
```

```python
#finding null value 
df['Summary'].isnull().sum()
df
```

```python
#convert Score column into pos & neg     ,, for compression with Summary rating score
df['rating_score_column'] = df['Score'].apply(lambda c: 'pos' if c > 2 else 'neg')
df.head(10)
```

```python
sid.polarity_scores(df.loc[0]['Summary'])
```

```python
#convert Summary into "positive" & "negative" value
df['sum_scores'] = df['Summary'].apply(lambda Summary: sid.polarity_scores(Summary))
df.head()
```

```python
#compund of summary sum_score
df['compound'] = df['sum_scores'].apply(lambda score_dict: score_dict['compound'])
df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >= 0 else 'neg')
df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#classification_report between 'rating_score_column' & 'comp_score'
print(classification_report(df['rating_score_column'],df['comp_score']))
```

```python
#confusion_matrix   between 'rating_score_column' & 'comp_score'
print(confusion_matrix(df['comp_score'],df['rating_score_column']))
```

```python
#checking accuracy_score   between 'rating_score_column' & 'comp_score'  
accuracy_score(df['comp_score'],df['rating_score_column'])
```
